```python
# 2b

def makeCategory(x):
  if x == 'Below $50':
    return 'Cheap'
  elif x == "$51-100":
    return 'Not Cheap' 
  elif x == "$101-200":
    return 'Expensive'
  else: 
    return 'Very Expensive'

df2 = pd.read_csv('/content/open-rice.csv')


df2['price_category'] = df2['price_range'].apply(makeCategory)

df2['price_category'].head(10)
```

```python
# 2c

def totalDislike():

  df2 = pd.read_csv('/content/open-rice.csv')
  a = df2.head(10)
  return a.groupby('price_range')['dislikes'].sum()
  

totalDislike()
```

```python
# 2d

def totalBookmarks():
  df2 = pd.read_csv('/content/open-rice.csv')
  a = df2.head(10)

  c = a[(a['food_type'] == 'Hong Kong Style')]
  return c['bookmarks'].sum()
  
totalBookmarks()
```

```python
# 2e

def extractReview():
  df2 = pd.read_csv('/content/open-rice.csv')
  a = df2.head(10)

  b = list(a['number_of_reviews'])
  c = {}
  n=0
  for i in b:
    c[n]= int(i[1:4])
    n += 1

  df3 = pd.Series(c , name='number_of_reviews')


  return df3

 
extractReview()

```
